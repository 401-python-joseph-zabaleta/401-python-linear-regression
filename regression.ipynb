### Importing Dependencies, Creating DataFrames with Panda from local Files:

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

---
### Displaying the first five rows, description, and info

```python
pdf = pd.read_csv("./data/practice2.csv")
print(f"practice data shape = {pdf.shape}")
```

```python
pdf.head(5)
```

```python
pdf.info()
```

---
### Show a list of the columns:

```python
print(pdf.columns)
```

---
### Selecting the Values:

```python
X = pdf['deneyim'].values
y = pdf['maas'].values
```

---
### Displaying the X-Axis and reshape it from 1D to 2D:

```python
X
```

```python
x_shaped = X.reshape(-1, 1)
```

---
### Split the data into Train and test sets using a Regression Model

```python
x_train, x_test, y_train, y_test = train_test_split(x_shaped,y,train_size=0.8,test_size=0.2,random_state=100)
print(f"X_train shape {x_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shap {x_test.shape}")
print(f"y_test shape {y_test.shape}")
print(y_test)
```

---
### Displaying Training Data:

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='red')
plt.xlabel('X-Axis: Miles Ran')
plt.ylabel('Y-Axis: Calories Burned')
plt.title('Training Data')
plt.show()
```

---
### Training Accuracy vs Test Accuracy:

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
y_predict = lm.predict(x_test)
print(f"Train accuracy {round( lm.score(x_train, y_train) * 100, 2)} %")
print(f"Test accuracy {round( lm.score(x_test,y_test) * 100, 2)} %")
```

---
### Plot our test Data:

```python
plt.scatter(x_train, y_train, color='red')
plt.plot(x_test, y_predict)
plt.xlabel('X-Axis: Miles Ran')
plt.ylabel('Y-Axis: Calories Burned')
plt.title('Practice Data Plot')
plt.show()
```
